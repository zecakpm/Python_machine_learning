```python
import pandas as pd
# index_col = 0, usind the first column as the index
df = pd.read_csv("/Users/joseformiga/Desktop/Python/pandas/diamonds.csv", index_col=0)
df.head()
```

```python
#cheching data, transformation needed to all non-numerical values
df["clarity"].unique()
```

```python
#will assign a number to each of non numerical values
#One way to do if the values are equally important
#df["cut"].astype("category").cat.codes
cut_class_dict = {"Fair":1,"Good":2,"Very Good":3,"Premium":4,"Ideal":5}
clarity_dict = {"I1":1,"SI2":2,"SI1":3,"VS2":4,"VS1":5,"VVS2":6,"VVS1":7,"IF":8}
color_dict = {"J":1,"I":2,"H":3,"G":4,"F":5,"E":6,"D":7}


df["cut"] = df["cut"].map(cut_class_dict)
df["clarity"] = df["clarity"].map(clarity_dict)
df["color"] = df["color"].map(color_dict)
df.head
```

```python
import sklearn
from sklearn import svm, preprocessing

#shuffle the dataset
df = sklearn.utils.shuffle(df)

#drop the column we want to predict, this case price
x = df.drop("price", axis=1).values
# scale your date, bring all values between 0 and 1 
x = preprocessing.scale(x)
#assign the variable we want to predict to y
y = df["price"].values

#training sample
test_size = 200

#using all, less than the 200 values
x_train = x[:-test_size]
y_train = y[:-test_size]

#using the last 200 values
x_test = x[-test_size:]
y_test = y[-test_size:]

#default for classifier clf 

clf = svm.SVR(kernel="linear")
#running the training
clf.fit(x_train, y_train)

```

```python
#How good the classifier is?
clf.score(x_test, y_test)
```

```python
#loop to show contrast the prices predicted and actual prices
for x,y in zip(x_test, y_test):
    print(f"Model: {clf.predict([x])[0]}, Actual:{y}")
```

```python
#other kernel type
clf = svm.SVR(kernel="rbf")
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```
